### Finding the trend of new airbnb listings in Seattle

```python
import pandas as pd

df = pd.read_csv('listings.csv')

df=df[['host_since']]
df.head()
```

```python
# Convert 'host_since' column to datetime format
df['host_since'] = pd.to_datetime(df['host_since'])

# Extract year and month from the 'host_since' column
df['year_month'] = df['host_since'].dt.to_period('M')

# df.head()
```

```python
# Group by the 'year_month' and count the number of entries
monthly_counts = df.groupby('year_month').size()

# Convert the result to DataFrame for better readability
monthly_counts_df = monthly_counts.reset_index(name='count')

# Display the result
print(monthly_counts_df)
```

```python
number = monthly_counts_df[['count']]
print(number)
```

```python
# Taking only the count column
number = monthly_counts_df[['count']]
print(number)
```

```python
# Plot the results
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

monthly_counts.plot(kind='bar', figsize=(10, 6))

# Plot the trend line
# Prepare data for trend line fitting
X = np.arange(len(number)).reshape(-1, 1)  # Convert to 2D array
y = number.values

# Fit the model
model = LinearRegression()
model.fit(X, y)
trend_line = model.predict(X)

# Plot the trend line
plt.plot(number.index, trend_line, color='r', linestyle='--', linewidth=2, label='Trend Line')

plt.title('New Airbnb Listings in Seattle (By Month)')
plt.xlabel('Month')
plt.ylabel('Number of New Listings')
plt.xticks(rotation=90, fontsize=8)
plt.grid(True)
plt.show()
```

### Finding the trend of total airbnb visitors to Seattle
